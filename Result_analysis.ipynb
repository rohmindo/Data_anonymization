```python
import pandas as pd
df = pd.read_excel('best_result.xlsx')
df.drop(['Unnamed: 0'], axis = 1, inplace = True)
```

```python
classname=[]
value=[]
classname=list(df.columns[5:])
```

```python
show(classname)
```

```python
def show(classname) :
  #ls = list(df.loc[3][5:])
  for j in range(4) :
    ls=list(df.loc[j][5:])
    example_tuples=list(zip(classname,ls))
    example_tuples.sort(key = lambda element : element[1])
    data=[]
    name=[]
    for i in range(15) :
    #print(example_tuples[i][1])
      data.append(example_tuples[i][1])
      name.append(example_tuples[i][0])
    import matplotlib.pyplot as plt
    plt.figure(figsize=(13, 5))
    plt.bar(name, data)
    plt.title('Test Accuracy')
    if j==0 : 
      plt.xlabel('Morphology')
      print('sdf')
    elif j==1 :
      plt.xlabel('Blocking & Merging')
    elif j==2 :
      plt.xlabel('Gaussian')
    else :
      plt.xlabel('Original')
    plt.ylabel('Accuracy')
    plt.ylim(0.7,1)
   # plt.subplot(4,1,i+1)
```
