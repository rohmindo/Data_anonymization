```python
import cv, cv2
import numpy as np

img1 = cv2.imread('myimage.jpg', cv2.CV_LOAD_IMAGE_GRAYSCALE)
h, w = img1.shape[:2]
vis0 = np.zeros((h,w), np.float32)
vis0[:h, :w] = img1
vis1 = cv2.dct(vis0)
img2 = cv.CreateMat(vis1.shape[0], vis1.shape[1], cv.CV_32FC3)
cv.CvtColor(cv.fromarray(vis1), img2, cv.CV_GRAY2BGR)

cv.SaveImage('output.jpg', img2)
```

```python
image = cv2.imread('tests.jpg') #Read the picture,
img=image.copy()
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.title('original')
plt.show()

img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
img1 = img.astype('float') #Convert uint8 to float type



img_dct = cv2.dct(img1) #Perform discrete cosine transform

img_dct_log = np.log(abs(img_dct)) #do log processing
height, width = img_dct_log.shape
for j in range( 0, height):
    for i in range(width):
        if img_dct_log[j,i]>4 :
          image[j][i]=0 

        #img_dct_log[j,i] = 0

 
 #display
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
plt.title('combine')
plt.show()

plt.imshow(img_dct_log)
plt.title('dct transformed')
 
plt.show()
```
