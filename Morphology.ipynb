```python
import os
root_dir = '/content/drive/MyDrive/mindoRoh_intern/mydataset/train' # 디렉토리
```

```python
##파일 디렉토리만들기
for (root, dirs, files) in os.walk(root_dir):
    if len(files) > 0:
      #새롭게만들 dataset폴더이름
      temp=root.replace('train','Mov2_train')
      print(temp)
      os.mkdir(temp)
```

```python
#파일 저장하기
for (root, dirs, files) in os.walk(root_dir):
    if len(files) > 0:
        for file_name in files:
                img_path = root + '/' + file_name
                image = cv2.imread(img_path)
                kernel_size_row = 7
                kernel_size_col = 7
                kernel = np.ones((7, 7), np.uint8)
                erosion_image = cv2.erode(image, kernel, iterations=2)  #// make erosion image
                dilation_image = cv2.dilate(image, kernel, iterations=2)  #// make dilation image
                #closed
                output_image=cv2.erode(dilation_image, kernel, iterations=1)

                output_path=img_path.replace('train','Mov2_train')
                print(output_path)
                cv2.imwrite(output_path,output_image)
                # 경로에서 \를 모두 /로 바꿔줘야함
                #img_path = img_path.replace('\\', '/') # \는 \\로 나타내야함         
                #img_path_list.append(img_path)
```
