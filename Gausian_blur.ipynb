```python
import os
root_dir = '/content/drive/MyDrive/mindoRoh_intern/mydataset/train' # 디렉토리
```

```python
##파일 디렉토리만들기
for (root, dirs, files) in os.walk(root_dir):
    if len(files) > 0:
      #새롭게만들 dataset폴더이름
      temp=root.replace('train','Gaussian_train')
      print(temp)
      os.mkdir(temp)
```

```python
#파일 저장하기
for (root, dirs, files) in os.walk(root_dir):
    if len(files) > 0:
        for file_name in files :
                img_path = root + '/' + file_name
                image = cv2.imread(img_path)
                output_image = cv2.GaussianBlur(image, (0, 0), 7)
                output_path=img_path.replace('train','Gaussian_train')
                print(output_path)
                cv2.imwrite(output_path,output_image)
```
