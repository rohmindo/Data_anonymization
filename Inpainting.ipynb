```python
import cv2

import numpy as np
import matplotlib.pyplot as plt
import os
root_dir = '/content/drive/MyDrive/mindoRoh_intern/mydataset/train' # 디렉토리
```

```python
#파일 저장하기
for (root, dirs, files) in os.walk(root_dir):
    if len(files) > 0:
        for file_name in files:
                img_path = root + '/' + file_name
                image = cv2.imread(img_path)
                height,width,ch = image.shape
                #block size 지정
                b_s=12 
# mask 이미지 및 missing part가지는 이미지 4장 정의
                image_missing=[]
                mask_img=[]
                final_img=np.zeros((height,width,3),dtype=np.uint8)
                
                for i in range(4) :
                  image_missing.append(image.copy())
                  mask_img.append(np.zeros((height,width,1),dtype=np.uint8))  
                
                for i in range(((height-1)//b_s)+1):
                  i_1=i*2
                  for j in range(((width-1)//b_s)+1):
                    j_1=j*2
                    image_missing[0][b_s*i_1:b_s*i_1+b_s,b_s*j_1:b_s*j_1+b_s]=255
                    image_missing[1][b_s*(i_1+1):b_s*(i_1+1)+b_s,b_s*j_1:b_s*j_1+b_s]=255
                    image_missing[2][b_s*i_1:b_s*i_1+b_s,b_s*(j_1+1):b_s*(j_1+1)+b_s]=255
                    image_missing[3][b_s*(i_1+1):b_s*(i_1+1)+b_s,b_s*(j_1+1):b_s*(j_1+1)+b_s]=255
                    mask_img[0][b_s*i_1:b_s*i_1+b_s,b_s*j_1:b_s*j_1+b_s]=255
                    mask_img[1][b_s*(i_1+1):b_s*(i_1+1)+b_s,b_s*j_1:b_s*j_1+b_s]=255
                    mask_img[2][b_s*i_1:b_s*i_1+b_s,b_s*(j_1+1):b_s*(j_1+1)+b_s]=255
                    mask_img[3][b_s*(i_1+1):b_s*(i_1+1)+b_s,b_s*(j_1+1):b_s*(j_1+1)+b_s]=255
                
                #inpainting
                dst=[]
                for i in range(4):
                  dst.append(cv2.inpaint(image_missing[i],mask_img[i],5,cv2.INPAINT_TELEA))

                for i in range(((height-1)//b_s)+1):
                  i_1=i*2
                  for j in range(((width-1)//b_s)+1):
                    j_1=j*2
                    final_img[b_s*i_1:b_s*i_1+b_s,b_s*j_1:b_s*j_1+b_s]=dst[0][b_s*i_1:b_s*i_1+b_s,b_s*j_1:b_s*j_1+b_s]
                    final_img[b_s*(i_1+1):b_s*(i_1+1)+b_s,b_s*j_1:b_s*j_1+b_s]=dst[1][b_s*(i_1+1):b_s*(i_1+1)+b_s,b_s*j_1:b_s*j_1+b_s]
                    final_img[b_s*i_1:b_s*i_1+b_s,b_s*(j_1+1):b_s*(j_1+1)+b_s]=dst[2][b_s*i_1:b_s*i_1+b_s,b_s*(j_1+1):b_s*(j_1+1)+b_s]
                    final_img[b_s*(i_1+1):b_s*(i_1+1)+b_s,b_s*(j_1+1):b_s*(j_1+1)+b_s]=dst[3][b_s*(i_1+1):b_s*(i_1+1)+b_s,b_s*(j_1+1):b_s*(j_1+1)+b_s]

                final_img_show=cv2.cvtColor(final_img,cv2.COLOR_BGR2RGB)
                gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

                #feature pixel 더해서 선명하게
                orb = cv2.ORB_create(nfeatures=1500)
                keypoints_orb, descriptors = orb.detectAndCompute(gray, None)
                for i in range(len(keypoints_orb)) :
                  x=int(keypoints_orb[i].pt[0])
                  y=int(keypoints_orb[i].pt[1])
                  final_img[y-1:y,x-1:x]=image[y-1:y,x-1:x]
                final_img_save=cv2.cvtColor(final_img,cv2.COLOR_BGR2RGB)

                final_img_save=cv2.cvtColor(final_img_save,cv2.COLOR_RGB2BGR)
                #cv2.imwrite("./image_block/feature_add.png", final_img_save)

                output_path=img_path.replace('train','Blocking_train')
                print(output_path)
                cv2.imwrite(output_path,final_img_save)
```
